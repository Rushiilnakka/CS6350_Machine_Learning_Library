```python
# Create a dictionary to map labels to numerical values
label_mapping = {}

# select all categorical variables
df_categorical_data = df.select_dtypes(include=['object'])

# Encode categorical variables using a custom label encoding
for column in df_categorical_data.columns:
    unique_values = df_categorical_data[column].unique()
    for i, value in enumerate(unique_values):
        label_mapping[value] = i
        df_categorical_data[column] = df_categorical_data[column].replace(value, i)

# Now, df_categorical_data contains the encoded categorical variables
df_categorical_data.head()
```

```python
df = df.drop(df_categorical_data.columns, axis=1)
df = pd.concat([df, df_categorical_data], axis=1)
df.head()
```

```python
df.info()
```

```python
# here the target variable is the income column, it can aslo be converted to categorical
df['income>50K'] = df['income>50K'].astype('category')
```

```python
# List of numerical and categorical columns
numerical_columns = ['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']
categorical_columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']

# Replace missing values ('?') with median for numerical columns
for col in numerical_columns:
    df_test[col] = df_test[col].replace('?', np.nan)

   
    df_test[col] = df_test[col].astype(float)


    df_test[col] = df_test[col].fillna(df_test[col].median())

# Replace missing values ('?') with mode for categorical columns
for col in categorical_columns:
   
    df_test[col] = df_test[col].replace('?', np.nan)
    df_test[col] = df_test[col].fillna(df_test[col].mode()[0])
```

```python
from sklearn import preprocessing


# encode categorical variables using Label Encoder

# select all categorical variables
df_categorical = df_test.select_dtypes(include=['object'])
df_categorical.head()
```

```python
# apply Label encoder to df_categorical

le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()
```
